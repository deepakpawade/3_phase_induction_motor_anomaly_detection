This code generates a test signal consisting of two sine waves at 10 Hz and 20 Hz, respectively. It then applies a 4th order Butterworth high-pass filter with a cut-off frequency of 15 Hz to the signal using the signal.butter() function. The signal.filtfilt() function is used to apply the filter to the signal. Finally, the original and filtered signals are plotted using the matplotlib library.

```python
import numpy as np
from scipy import signal

# Create a test signal
t = np.linspace(0, 1, 1000, False)
sig = np.sin(2 * np.pi * 10 * t) + np.sin(2 * np.pi * 20 * t)

# Define the high-pass filter
fc = 15  # Cut-off frequency
b, a = signal.butter(4, fc/(1000/2), 'highpass')

# Apply the filter to the signal
filtered_signal = signal.filtfilt(b, a, sig)

# Plot the original and filtered signals
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(t, sig)
ax1.set_title('Original Signal')
ax2.plot(t, filtered_signal)
ax2.set_title('High-pass Filtered Signal')
plt.show()
```
