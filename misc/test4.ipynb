```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Load all 317 files into a single DataFrame
# Read the data in chunks
datafolder = 'data/s3Files/'
data_files = os.listdir(datafolder)
df = pd.DataFrame()
for file in data_files:
    # print(file)
    df = pd.concat([df,pd.read_csv(datafolder+file,header=None,sep=',')])
df.drop(columns=[0,4],inplace=True)
df.dropna(inplace=True)
df.columns = ['current_1', 'current_2', 'current_3']
```

```python
df.shape
```

```python
df1 = df.reset_index(drop=True)
```

```python
df1
```

```python

# Calculate rolling mean and standard deviation over a window of 10000 time steps
window_size = 60
rolling_mean = df1.rolling(window_size).mean()
rolling_std = df1.rolling(window_size).std()

# Plot the original data along with the rolling mean and standard deviation

# fig, ax = plt.subplots()
# df.plot(ax=ax, label='Original')
# rolling_mean.plot(ax=ax, label='Rolling Mean')
# rolling_std.plot(ax=ax, label='Rolling Std')
# ax.legend()
# plt.show()
```

```python
df1.plot( label='Original')
plt.show()
```

```python
rolling_mean.plot( label='Rolling Mean')
plt.show()
```
