```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load the current data from a CSV file
df = pd.read_csv('data/s3Files/data12.txt',header=None,sep=',')
df.drop(columns=[0,4],inplace=True)
df.dropna(inplace=True)
df.columns = ['current_1', 'current_2', 'current_3']

```

```python
# Calculate the FFT of the data for each of the three currents
fft_data = np.fft.fft(df, axis=0)
# calculate the FFT of the data along the time axis (axis=0),
# resulting in an array of complex values representing the frequency spectrum of the data.
```

```python

# Calculate the frequency axis in Hz
n_samples = len(df)
sampling_rate = 10000 # assume a sampling rate of 10 kHz
freq_axis = np.fft.fftfreq(n_samples, 1/sampling_rate)
#  calculate the corresponding frequency axis in Hz, 
#  based on the sampling rate of 10 kHz assumed in this example.

# Plot the frequency spectrum for the first current
plt.plot(freq_axis, np.abs(fft_data[:,0]))
plt.plot(freq_axis, np.abs(fft_data[:,1]))
plt.plot(freq_axis, np.abs(fft_data[:,2]))

plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```
